```python
#Answer of 2
class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def extract_digits(head):
    # function to extract digits from linked list and return as a list
    digits = []
    while head:
        digits.append(head.val)
        head = head.next
    return digits

def LL_add(u, v):
    # extract digits from linked lists u and v
    digits1 = extract_digits(u.next)
    digits2 = extract_digits(v.next)
    
    # reverse the digits to form two new linked lists
    l1 = Node(digits1.pop())
    curr = l1
    while digits1:
        curr.next = Node(digits1.pop())
        curr = curr.next
    
    l2 = Node(digits2.pop())
    curr = l2
    while digits2:
        curr.next = Node(digits2.pop())
        curr = curr.next
    
    # add the digits from l1 and l2, maintaining carry
    carry = 0
    l3 = None
    while l1 or l2 or carry:
        val = carry
        if l1:
            val += l1.val
            l1 = l1.next
        if l2:
            val += l2.val
            l2 = l2.next
        carry, val = divmod(val, 10)
        if not l3:
            l3 = Node(val)
            curr = l3
        else:
            curr.next = Node(val)
            curr = curr.next
    
    # reverse the new linked list l3 to get the final result
    result = Node(0)
    curr = l3
    while curr:
        temp = curr.next
        curr.next = result.next
        result.next = curr
        curr = temp
    
    return result.next
u = Node(0, Node(1, Node(2, Node(3, Node(4)))))
v = Node(0, Node(5, Node(6, Node(7, Node(8)))))

result = LL_add(u, v)
while result:
    print(result.val, end="->")
    result = result.next
```

```python
#Answer of 3
class Node:
    def __init__(self, label, next=None):
        self.label = label
        self.next = next

def eliminate_monkey(curr, m):
    # base case: only one monkey remains
    if curr.next == curr:
        return curr
    # eliminate the m-th monkey
    for i in range(m-2):
        curr = curr.next
    curr.next = curr.next.next
    # recurse on the remaining monkeys
    return eliminate_monkey(curr.next, m)

def find_monkey_king(num_monkeys, m):
    # create a circular linked list with num_monkeys nodes
    head = Node(0)
    curr = head
    for i in range(1, num_monkeys):
        curr.next = Node(i)
        curr = curr.next
    curr.next = head

    # eliminate every m-th monkey until only one monkey remains
    winner = eliminate_monkey(head, m)

    # return the label of the remaining monkey
    return winner.label

# Example usage:
num_monkeys = int(input("Enter total number of monkeys in a group: "))
m = int(input("Enter m value: "))
king = find_monkey_king(num_monkeys, m)
print("The king will be", king)
```

```python
#Answer of 1
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

def srt_LL(head):
    if not head:
        return None
    
    # Create an empty dictionary to count the occurrences of each character
    char_counts = {}
    
    # Traverse the linked list and count the characters
    curr = head
    while curr:
        char = curr.val
        if char not in char_counts:
            char_counts[char] = 1
        else:
            char_counts[char] += 1
        curr = curr.next
    
    # Create a new linked list with the characters sorted
    new_head = None
    for char in sorted(char_counts.keys()):
        count = char_counts[char]
        for i in range(count):
            new_node = Node(char)
            if not new_head:
                new_head = new_node
                curr = new_head
            else:
                curr.next = new_node
                curr = curr.next
    return new_head
# Define the linked list
head = Node('D')
head.next = Node('A')
head.next.next = Node('C')
head.next.next.next = Node('A')
head.next.next.next.next = Node('G')



# Sort the linked list
new_head = srt_LL(head)

# Print the sorted linked list
curr = new_head
while curr:
    print(curr.val, end=' ')
    curr = curr.next
print()
```
